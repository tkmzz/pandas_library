```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%pip install plotly
%pip install cufflinks
%pip install chart-studio
import chart_studio.plotly as py
import cufflinks as cf
import plotly.graph_objects as go
import plotly.express as px
```

```python
sales = pd.read_excel("varejo.xlsx")
sales.head()
```

```python
sales.info()
```

```python
sales.groupby("idcanalvenda").idcompra.nunique()
```

```python
sales["idcanalvenda"] = sales["idcanalvenda"].str.replace("APP", "Aplicativo")
sales.groupby("idcanalvenda").idcompra.nunique()
```

```python
sales.groupby("bandeira").idcompra.nunique()
```

```python
sales.groupby("Nome_Departamento").idcompra.nunique()
```

```python
sales.groupby("estado").idcompra.nunique()
```

```python
sales["Nome_Departamento"] = sales["Nome_Departamento"].str.replace(" ", "_")
sales.groupby("Nome_Departamento").idcompra.nunique()
```

```python
sales.info()
```

```python
sales[sales["estado"].isnull()]
```

```python
sales_not_null = sales.dropna()
sales_not_null.info()
```

```python
sales["estado"].fillna("MS", inplace=True)
```

```python
sales.info()
```

```python
avg_price = sales["Preço"].mean()
```

```python
sales["Preço"].fillna(avg_price, inplace=True)
```

```python
sales.info()
```

```python
sales[["Preço","Preço_com_frete"]].describe()
```

```python
wrong_price = sales.query("Preço > Preço_com_frete")
right_sales = sales.query("Preço < Preço_com_frete")
```

```python
right_sales.describe()
```

```python
right_sales.head()
```

```python
right_sales.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
right_sales.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(right_sales.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
right_sales.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
right_sales["mes"] = right_sales["Data"].dt.month
```

```python
right_sales.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

```python
client = pd.read_excel("cliente_varejo.xlsx")
client.head()
```

```python
right_sales.head()
```

```python
client[["idade", "renda"]].describe()
```

```python
client.info()
```

```python
client = client.astype({"renda":'float'})
client.info()
```

```python
client_sales = right_sales.merge(client, how='left', on='cliente_Log')
client_sales.head()
```

```python
agg_idcanal_sales = round(client_sales.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_sales
```

```python
agg_idade_bandeira = round(client_sales.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')
plt.ylabel('Média de idade')
plt.title('Idade média por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_sales['idcanalvenda'], agg_idcanal_sales['renda'], color='green')
plt.ylabel('Renda média')
plt.title('Renda média por canal de venda', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
sales_by_date = right_sales.groupby('Data').idcompra.nunique().reset_index()
sales_by_date
```

```python
fig = plt.figure(figsize=(20,12))
plt.xlabel('Data da venda')
plt.ylabel('Qtd de vendas')
x = sales_by_date['Data']
y = sales_by_date['idcompra']
plt.plot(x,y, color='green', linewidth='2')
plt.show()
```

```python
agg_dept_price = round(right_sales.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dept_price
```

```python
fig = plt.figure(figsize=(30,20))
plt.bar(agg_dept_price['Nome_Departamento'], agg_dept_price['Preço_com_frete'], color='green')
plt.ylabel('Preço médio')
plt.title('Preço médio por departamento', loc='center', fontsize=18, fontweight='bold', color='black');
```
